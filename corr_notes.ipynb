```python
import pandas as pd
```

```python
df_1=pd.merge(pd.read_csv("FF_Processed_Data_a.csv"),(pd.read_csv("FF_Processed_Data_b.csv")),on='idnum')
print('pair 1...')
df_2=pd.merge((pd.read_csv("FF_Processed_Data_c.csv")),(pd.read_csv("FF_Processed_Data_d.csv")),on='idnum')
print('pair 2...')
df_3=pd.merge(df_1,df_2,on='idnum')
print('almost...')
df=pd.merge(df_3,(pd.read_csv("FF_Processed_Data_e.csv")),on='idnum')
df
```

```python
reps=list(df.keys())
for x in reps:
    for y in reps:
        if abs(df[x].corr(df[y], min_periods=100))>0.8 and x!=y and x[2]!='n' and x[3]!='i'and y[3]!='i'and y[2]!='n':
            print(x+' and '+y+' are corr by '+str(df[x].corr(df[y])))
print('Done!')
```
