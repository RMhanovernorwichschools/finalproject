```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def corr (x, y):
    print(x+' and '+y+' are corr by '+str(df[x+'_int'].corr(df[y+'_int'])))

df_1=pd.merge(pd.read_csv("FF_Processed_Data_a.csv"),(pd.read_csv("FF_Processed_Data_b.csv")),on='idnum')
print('pair 1...')
df_2=pd.merge((pd.read_csv("FF_Processed_Data_c.csv")),(pd.read_csv("FF_Processed_Data_d.csv")),on='idnum')
print('pair 2...')
df_3=pd.merge(df_1,df_2,on='idnum')
print('pair 3...')
df_4=pd.merge(df_3,(pd.read_csv("FF_Processed_Data_e.csv")),on='idnum')
print('almost...')

df=df_4=pd.merge(df_4,(pd.read_csv("FF_Processed_Data_addi.csv")),on='idnum')
print('Done!')
```

```python
zoo=[]
zoo_up=[]
zoo_lo=[]
for x in range(9):
    series=df.k6d1b_int[df.cp6kids_int==x]
    if series.count()>20:
        zoo.append(series.mean())
        zoo_up.append(series.mean()+series.std())
        zoo_lo.append(series.mean()-series.std())

fig, ax = plt.subplots()

index = np.arange(len(zoo))
bar_width = 0.3

opacity = 0.6

rects2 = ax.bar(index, zoo_up, 0.1,
                alpha=0.3, color='r',
                label='1 std up')
rects1 = ax.bar(index+0.2, zoo, bar_width,
                alpha=opacity, color='b',
                label='Comforts Others')
rects3 = ax.bar(index + bar_width+0.1, zoo_lo, 0.1,
                alpha=0.3, color='r',
                label='1 std down')

ax.set_xlabel('Number of Kids')
ax.set_ylabel('Score')
plt.ylim([1.0, 3.0])
ax.set_title("Demonstration of no Significant Correlation")
ax.set_xticks(index + bar_width / 2)
plt.text(4.8, 2, '''A score of 1 
corresponds to 
the response 
"Not True", 2 
to "Sometimes 
True", and 3 
to "Often True"''')
ax.set_xticklabels(('', '2', '3', '4', '5', '6', '7', '8', '9'))
ax.legend()

fig.tight_layout()
corr('cp6kids','k6d1b')
plt.show()
```

The conclusion suggested by the graph above is that *the nonresident parent, when a parent is living apart from a child, tends to be the biological father.*

This has been observed in other ways and supports previous beliefs, so it is likely a valid conclusion. It also shows, given that the nonresident parent is usually the father, that the primary caregiver tends to think the child has seen his/her father the same number of times the child reports doing so.
